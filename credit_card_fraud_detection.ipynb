```python
import numpy as np 
import pandas as  pd
```

```python
# upsampling
```

```python
df_minority = df[df['Class']==1]
df_majority = df[df['Class']==0]
```

```python
df_majority = df_majority.sample(100000,random_state=2)
```

```python
df_majority.shape # resizing the datapoints to 1Lakhs
```

```python
len(df_majority)
```

```python
## performing upsampling
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsample = resample(df_minority,
                        replace=True ,
                       n_samples = len(df_majority), # want to match df_majority
                       random_state =2)
```

```python
df_minority_upsample.shape
```

```python
df_minority_upsample
```

```python
df_minority_upsample['Class'].value_counts()
```

```python
# after upsampling our new dataframe
new_df= pd.concat([df_minority_upsample,df_majority])
```

```python
new_df.shape
```

```python
new_df = new_df.reset_index(drop=True)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

```python
# our 50% class =1 and 50% class =0 data is perfectly balanced
```

```python
# checking the null values
```

```python
new_df.isnull().sum()
```

```python
# there is no null values
```

```python
X=new_df.drop(labels=["Class"],axis=1)
```

```python
X.shape
```

```python
X
```

```python
y=new_df["Class"]
```

```python
y.shape
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#we did train and test split before only why because we dont want data leakage
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=50)
```

```python
#preprocessing steps
from sklearn.preprocessing import StandardScaler
train_scaler=StandardScaler()
test_scaler=StandardScaler()
```

```python
X_train_scaled=train_scaler.fit_transform(X_train)
```

```python
X_test_scaled=train_scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create an instance of the RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=10, random_state=42)

# Fit the model to your data
rf_classifier.fit(X_train_scaled, y_train)

# Make predictions
predictions = rf_classifier.predict(X_test)
```

```python
y_pred = predictions
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
